```python
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
```

```python
data_europe = pd.read_csv('gapminder_gdp_europe.csv', index_col = 'country')
years = data_europe.columns.str.strip('gdpPercap_')
data_europe.columns = years.astype(int)


data_europe
```

## Print the statistics of the DataFrame (mean, std, quartiles) 
### Use the .describe()


```python
data_europe.describe()
```

## Create a mask where you only show countries that have gpd greater than 10000


```python
mask = data_europe > 10000
print(data_europe[mask])
```

## Pick 3 countries and make 1 scatter plot year vs GPD of the 3 countries in the subset data
## Add labels, legend, title

```python
years=data_europe.columns
plot_set = data_europe.loc['Czech Republic':'Finland']
plot_set.T.plot()
d_czech = data_europe.loc['Czech Republic']
d_denmark = data_europe.loc['Denmark']
d_finland = data_europe.loc['Finland']
mp.scatter(years, d_czech)
mp.scatter(years, d_denmark)
mp.scatter(years, d_finland)
mp.xlabel('Year')
mp.ylabel('GDP')
```

# Extra interactive 3D plot 

```python
#From https://www.geeksforgeeks.org/three-dimensional-plotting-in-python-using-matplotlib/

#Make plot interactive
%matplotlib notebook  


fig = mp.figure()
 
# syntax for 3-D projection
ax = mp.axes(projection ='3d')
 
# defining all 3 axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
mp.show()

```
