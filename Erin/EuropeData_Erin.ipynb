```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_europe = pd.read_csv('gapminder_gdp_europe.csv', index_col = 'country')
years = data_europe.columns.str.strip('gdpPercap_')
data_europe.columns = years.astype(int)


data_europe
```

## Print the statistics of the DataFrame (mean, std, quartiles) 
### Use the .describe()


```python
data_europe.describe()
```

## Create a mask where you only show countries that have gpd greater than 10000


```python
mask = data_europe > 10000
print(mask)
```

## Pick 3 countries and make 1 scatter plot year vs GPD of the 3 countries in the subset data
## Add labels, legend, title

```python
Germany = data_europe.loc["Germany",:]
Italy = data_europe.loc["Italy",:]
Spain = data_europe.loc["Spain",:]

plt.scatter(years, Germany, c='k', label='Germany')
plt.scatter(years, Italy, c='g', label='Italy')
plt.scatter(years, Spain, c='r', label='Spain')
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP/Capita Over Time for 3 Selected Countries")
plt.legend(loc='best')
plt.show()
```

# Optional
### Explore the data set
### Make an extra plot 

```python
Ireland = data_europe.loc["Ireland",:]
UK = data_europe.loc["United Kingdom",:]

plt.scatter(years, Ireland, c='g', label='Ireland')
m_ir, b_ir = np.polyfit((years.values).astype(int), (Ireland.values).astype(int), 1)
plt.plot((years.values).astype(int), m_ir*(years.values).astype(int) + b_ir, c='k')

plt.scatter(years, UK, c='b', label='United Kingdom')
m_uk, b_uk = np.polyfit((years.values).astype(int), (UK.values).astype(int), 1)
plt.plot((years.values).astype(int), m_uk*(years.values).astype(int) + b_uk, c='r')

plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP/Capita Over Time in the British Isles")
plt.legend(loc='best')
plt.show()
```

# Extra interactive 3D plot 

```python
#From https://www.geeksforgeeks.org/three-dimensional-plotting-in-python-using-matplotlib/

#Make plot interactive
%matplotlib notebook  


fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining all 3 axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()

```
