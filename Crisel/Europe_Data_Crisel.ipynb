```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_europe = pd.read_csv('gapminder_gdp_europe.csv', index_col = 'country')
years = data_europe.columns.str.strip('gdpPercap_')
data_europe.columns = years.astype(int)


data_europe
```

```python
#Print the statistics of the DataFrame (mean, std, quartiles) 
#Use the .describe()
```

```python
data_europe.describe()
```

```python
# Create a mask where you only show countries that have gpd greater than 10000
# DataFrame where 
```

```python
europe_mask = data_europe > 10000
subset_europe = data_europe[europe_mask]
subset_europe
```

```python
# Pick 3 countries and make 1 scatter plot year vs GPD of the subset data
# Add labels, lengend, title
```

```python
plt.scatter(subset_europe.columns, subset_europe.loc['Spain'], label = 'Spain')
plt.scatter(subset_europe.columns, subset_europe.loc['Greece'], label = 'Greece')
plt.scatter(subset_europe.columns, subset_europe.loc['France'], label = 'France')
plt.legend(loc='best')
plt.title('Spain, Greece, France')
plt.ylabel('GDP')
plt.xlabel('Year')

```

```python
#From https://www.geeksforgeeks.org/three-dimensional-plotting-in-python-using-matplotlib/

#Make plot interactive
%matplotlib notebook  


fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining all 3 axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot geeks for geeks')
plt.show()

```
